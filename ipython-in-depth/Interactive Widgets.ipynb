# Interactive Javascript widgets: "micro-GUIs" in the browser

IPython 2.0 offers an `interact` function and decorator for interactive exploration. This is the highest-level API in our interactive JavaScript architecture.

## Image editing

In this example, we will perform some elementary transformations to an image and display the results with Matplotlib.  If you don't have matplotlib installed, you can still follow replacing the plotting of images with simply printing the arrays. In that case, skip the next cell.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np

i = np.load('figs/coffee.npy')
# replace with a 'print' call if you don't have matplotlib
plt.imshow(i);
```

Here is a function that can make simple adjustments to the RGB channels of an image:

```python
def edit_image(image, r=1.0, g=1.0, b=1.0):
    new_image = image.copy()
    new_image[:,:,0] = r*new_image[:,:,0]
    new_image[:,:,1] = g*new_image[:,:,1]
    new_image[:,:,2] = b*new_image[:,:,2]    
    # Display the figure with matplotlib
    plt.imshow(new_image)
    return new_image
```

Calling the function once, displays and returns the edited image:

```python
new_i = edit_image(i, r=2.8);
```

We can use `interact` to explore the parameter space of the processed image:

```python
from IPython.html.widgets import interact, fixed

lims = (0, 3, 0.05)
interact(edit_image, image=fixed(i), r=lims, g=lims, b=lims);
```

We can quickly interate through the visualize, interact, compute cycle.

## How does this work?

* The first argument to interact is a callable/function
* The keyword arguments to `interact` are "widget abbreviations"
* These "widget abbreviations" are converted to `Widget` instances
* These `Widgets` objects are Python objects that are automatically synchronized with JavaScript MVC objects running in the browser.
* `interact` simply calls its callable each time any widget changes state

```python
def f(x):
    print x
```

```python
interact(f, x=True);
```

```python
interact(f, x=(0,10,2));
```

```python
interact(f, x='Hello World');
```

```python
interact(f, x=dict(this=list, that=tuple, other=str));
```

## The @interactive decorator

```python
from IPython.html.widgets import interactive

@interactive
def g(x=True):
    print x
```
